### 1.2 문자열 (string)

텍스트 데이터를 다루는 자료형입니다. 따옴표(`'` 또는 `"`)로 감싸서 만듭니다.

```python
# 문자열 만들기
국가 = "Korea"
평가명 = 'PISA 2022'
print(국가, "-", 평가명)

# 문자열 합치기 (concatenation)
소개 = 국가 + "는 " + 평가명 + "에 참여했습니다."
print(소개)
```

```python
# 문자열의 유용한 기능들
텍스트 = "  Hello, PISA World!  "
print(텍스트.strip())       # 양쪽 공백 제거
print(텍스트.lower())       # 소문자로
print(텍스트.upper())       # 대문자로
print(텍스트.replace("PISA", "Data"))  # 치환
print(len(텍스트.strip()))  # 글자 수
```

### 1.3 불린 (boolean) - 참/거짓

데이터 분석에서 **조건을 거는 데** 핵심적으로 쓰입니다.

```python
# 비교 연산자 → 결과가 True 또는 False
print(527 > 472)    # True  (한국 > OECD 평균)
print(527 == 472)   # False (같은지 비교)
print(527 != 472)   # True  (다른지 비교)
print(527 >= 530)   # False
print("KOR" == "KOR")  # True
```

### 1.4 리스트 (list) - 여러 값을 순서대로 담기

데이터 분석에서 **여러 값을 한 번에 다루는** 가장 기본적인 구조입니다.

```python
# 리스트 만들기
과목들 = ["수학", "읽기", "과학"]
점수들 = [527, 515, 528]

print("과목:", 과목들)
print("점수:", 점수들)
print("과목 수:", len(과목들))
```

```python
# 인덱싱 - 특정 위치의 값 꺼내기 (0부터 시작!)
print(과목들[0])    # "수학" (첫 번째)
print(과목들[1])    # "읽기" (두 번째)
print(과목들[-1])   # "과학" (마지막)
```

```python
# 리스트에 값 추가/제거
과목들.append("창의적 사고")
print("추가 후:", 과목들)

과목들.remove("창의적 사고")
print("제거 후:", 과목들)
```

```python
# 슬라이싱 - 리스트의 일부분 잘라내기
숫자 = [10, 20, 30, 40, 50]
print(숫자[1:3])    # [20, 30] — 1번~2번 (3 미포함)
print(숫자[:3])     # [10, 20, 30] — 처음~2번
print(숫자[2:])     # [30, 40, 50] — 2번~끝
```

### 1.5 딕셔너리 (dictionary) - 이름표가 있는 데이터

리스트가 **순서**로 구분한다면, 딕셔너리는 **이름(키)**으로 구분합니다.

### 2.2 for문 - 반복 작업

같은 작업을 여러 번 할 때 사용합니다. 데이터 분석에서는 "모든 행에 대해 ~를 해라"에 해당.

```python
# 리스트의 모든 항목에 대해 반복
과목들 = ["수학", "읽기", "과학"]
점수들 = [527, 515, 528]

for 과목 in 과목들:
    print(f"{과목} 영역")
```

```python
# enumerate: 순서(인덱스)도 함께 가져오기
for i, 과목 in enumerate(과목들):
    print(f"{i+1}번째 과목: {과목}, 점수: {점수들[i]}")
```

```python
# for + if 조합: 조건에 맞는 것만 골라내기
점수들 = [527, 515, 528, 480, 390, 610]

고득점 = []
for 점수 in 점수들:
    if 점수 >= 500:
        고득점.append(점수)

print(f"500점 이상: {고득점}")
print(f"총 {len(고득점)}명")
```

```python
# 리스트 컴프리헨션 — 위 코드를 한 줄로!
고득점 = [점수 for 점수 in 점수들 if 점수 >= 500]
print(고득점)
```

### 2.3 함수 - 반복되는 작업을 이름 붙여 재사용

"이 계산을 여러 번 써야 할 것 같은데?" → 함수로 만들어두세요.

```python
# 함수 정의
def 성취수준(점수):
    """PISA 점수를 기준으로 성취 수준을 반환합니다."""
    if 점수 >= 625:
        return "Level 5-6 (최상위)"
    elif 점수 >= 553:
        return "Level 4 (상위)"
    elif 점수 >= 482:
        return "Level 3 (중상위)"
    elif 점수 >= 420:
        return "Level 2 (기본)"
    else:
        return "Level 1 이하 (기초 미달)"

# 함수 사용
print(성취수준(527))
print(성취수준(390))
print(성취수준(650))
```

```python
# 함수 + 반복문 조합
샘플점수 = [527, 390, 650, 480, 560]

for 점수 in 샘플점수:
    print(f"{점수}점 → {성취수준(점수)}")
```

### 3.5 데이터 필터링 — 조건에 맞는 행 골라내기

**데이터 분석의 핵심 중 핵심!** "~인 학생만 골라서 보고 싶다"를 코드로 표현하는 법.

```python
# 수학점수가 500점 이상인 학생만
고득점 = df[df["수학점수"] >= 500]
print(f"500점 이상: {len(고득점)}명")
고득점
```

```python
# 여러 조건 조합 (& = 그리고, | = 또는)
# 여학생이면서 수학 500점 이상
조건 = (df["성별"] == "여") & (df["수학점수"] >= 500)
df[조건]
```

```python
# 특정 값 목록에 포함되는지 (isin)
# ICT 학교 사용이 3 또는 4인 학생
df[df["ICT학교사용"].isin([3, 4])]
```
