```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'  # Colab 기본 폰트
# 한글 폰트 설정 (Colab)
!apt-get install -y fonts-nanum > /dev/null 2>&1
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

import warnings
warnings.filterwarnings('ignore')

print("✅ 준비 완료!")
```

```python
# 연습용 데이터 만들기
import numpy as np
np.random.seed(42)

n = 200
df = pd.DataFrame({
    '학생ID': [f'S{str(i).zfill(3)}' for i in range(1, n+1)],
    '성별': np.random.choice(['남', '여'], n),
    'ICT학교사용': np.random.choice([1,2,3,4,5], n, p=[0.1,0.2,0.35,0.25,0.1]),
    'ICT가정사용': np.random.choice([1,2,3,4,5], n, p=[0.05,0.15,0.3,0.3,0.2]),
    'ICT활용유형': np.random.choice(['수동적', '능동적', '혼합'], n, p=[0.3,0.3,0.4]),
})
# 수학점수: ICT 활용유형과 약한 상관관계를 가지도록
base = np.random.normal(500, 60, n)
df['수학점수'] = (base + 
    df['ICT활용유형'].map({'수동적': -20, '능동적': 25, '혼합': 5}) +
    df['ICT학교사용'] * 5 +
    np.random.normal(0, 20, n)).astype(int)

df.head()
```

---
## 1. 히스토그램 (Histogram) — 분포 확인

**"이 변수의 값들이 어떻게 퍼져 있는가?"**를 보는 그래프입니다.
가장 먼저 그려야 할 그래프예요.

```python
# 기본 히스토그램
plt.figure(figsize=(8, 5))
plt.hist(df['수학점수'], bins=20, color='steelblue', edgecolor='white')
plt.xlabel('수학 점수')
plt.ylabel('학생 수')
plt.title('수학 점수 분포')
plt.axvline(df['수학점수'].mean(), color='red', linestyle='--', 
            label=f"평균: {df['수학점수'].mean():.1f}")
plt.legend()
plt.show()
```

**읽는 법:**
- x축: 점수 구간
- y축: 해당 구간에 속하는 학생 수
- 빨간 점선: 평균값
- 종 모양이면 정규분포에 가까움
- 한쪽으로 치우쳐 있으면 편향(skew)이 있음

## 2. 막대그래프 (Bar Chart) — 그룹별 비교

**"그룹 간에 차이가 있는가?"**를 보는 그래프.

```python
# 성별에 따른 평균 수학점수
성별_평균 = df.groupby('성별')['수학점수'].mean()

plt.figure(figsize=(6, 5))
성별_평균.plot(kind='bar', color=['#4A90D9', '#E8636E'], edgecolor='white')
plt.xlabel('성별')
plt.ylabel('평균 수학 점수')
plt.title('성별에 따른 평균 수학 점수')
plt.xticks(rotation=0)

# 값 표시
for i, v in enumerate(성별_평균):
    plt.text(i, v + 2, f'{v:.1f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ICT 활용 유형별 평균 점수 — 이 프로젝트의 핵심 그래프!
유형_평균 = df.groupby('ICT활용유형')['수학점수'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
colors = ['#2ecc71', '#3498db', '#e74c3c']
유형_평균.plot(kind='bar', color=colors, edgecolor='white')
plt.xlabel('ICT 활용 유형')
plt.ylabel('평균 수학 점수')
plt.title('ICT 활용 유형에 따른 평균 수학 점수')
plt.xticks(rotation=0)

for i, v in enumerate(유형_평균):
    plt.text(i, v + 2, f'{v:.1f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n📊 능동적 활용이 수동적 활용보다 평균 {유형_평균['능동적'] - 유형_평균['수동적']:.1f}점 높습니다.")
```

## 3. 산점도 (Scatter Plot) — 두 변수의 관계

**"이 변수가 높으면 저 변수도 높은가?"**를 보는 그래프.
상관관계를 시각적으로 확인하는 데 가장 좋습니다.

```python
# ICT 학교사용 vs 수학점수
plt.figure(figsize=(8, 6))
plt.scatter(df['ICT학교사용'], df['수학점수'], 
            alpha=0.5,  # 투명도 (겹치는 점 확인용)
            color='steelblue',
            edgecolors='white', linewidth=0.5)
plt.xlabel('ICT 학교 사용 빈도')
plt.ylabel('수학 점수')
plt.title('ICT 학교 사용과 수학 점수의 관계')

# 추세선 추가
z = np.polyfit(df['ICT학교사용'], df['수학점수'], 1)
p = np.poly1d(z)
x_line = np.linspace(1, 5, 100)
plt.plot(x_line, p(x_line), "r--", alpha=0.8, label=f'추세선')
plt.legend()
plt.show()
```

```python
# 그룹별로 색을 다르게 — 더 많은 정보를 담기
plt.figure(figsize=(9, 6))

색상 = {'수동적': '#e74c3c', '능동적': '#2ecc71', '혼합': '#3498db'}

for 유형, 그룹 in df.groupby('ICT활용유형'):
    plt.scatter(그룹['ICT학교사용'], 그룹['수학점수'],
                alpha=0.5, color=색상[유형], label=유형, 
                edgecolors='white', linewidth=0.5, s=50)

plt.xlabel('ICT 학교 사용 빈도')
plt.ylabel('수학 점수')
plt.title('ICT 활용 유형별 학교 사용 빈도와 수학 점수')
plt.legend(title='활용 유형')
plt.show()
```

## 4. 박스플롯 (Box Plot) — 분포와 이상치를 한눈에

**"그룹 간 분포 차이를 비교할 때"** 히스토그램보다 깔끔합니다.

```python
# ICT 활용 유형별 수학 점수 분포 비교
plt.figure(figsize=(8, 6))

# 유형별로 데이터 분리
유형들 = ['수동적', '혼합', '능동적']
데이터 = [df[df['ICT활용유형'] == t]['수학점수'] for t in 유형들]

bp = plt.boxplot(데이터, labels=유형들, patch_artist=True,
                 boxprops=dict(facecolor='lightblue', color='steelblue'),
                 medianprops=dict(color='red', linewidth=2))

plt.ylabel('수학 점수')
plt.title('ICT 활용 유형별 수학 점수 분포')
plt.show()
```

**박스플롯 읽는 법:**
- 상자의 아래쪽 선: 25% 지점 (Q1)
- 상자 안의 빨간 선: 중앙값 (50%, 중간값)
- 상자의 위쪽 선: 75% 지점 (Q3)
- 수염(whisker): Q1 - 1.5×IQR ~ Q3 + 1.5×IQR
- 수염 밖의 점: **이상치 (outlier)**

**해석 팁:**
- 상자가 높으면 → 데이터가 넓게 퍼져 있음
- 중앙값이 높으면 → 전반적으로 점수가 높은 그룹
- 이상치가 많으면 → 극단적인 값이 있음

## 5. 파이 차트 (Pie Chart) — 비율 확인

```python
# ICT 활용 유형의 비율
유형_비율 = df['ICT활용유형'].value_counts()

plt.figure(figsize=(7, 7))
colors = ['#e74c3c', '#3498db', '#2ecc71']
plt.pie(유형_비율, labels=유형_비율.index, autopct='%1.1f%%',
        colors=colors, startangle=90,
        wedgeprops=dict(edgecolor='white', linewidth=2))
plt.title('ICT 활용 유형 비율')
plt.show()
```

## 6. 여러 그래프를 한 화면에 — subplot

분석 보고서에 넣을 때 유용합니다.

```python
# 2x2 레이아웃으로 4개 그래프를 한 번에
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# (1) 점수 분포
axes[0,0].hist(df['수학점수'], bins=20, color='steelblue', edgecolor='white')
axes[0,0].set_title('수학 점수 분포')
axes[0,0].set_xlabel('점수')

# (2) 성별 평균
성별_평균 = df.groupby('성별')['수학점수'].mean()
axes[0,1].bar(성별_평균.index, 성별_평균.values, color=['#4A90D9', '#E8636E'])
axes[0,1].set_title('성별 평균 점수')

# (3) 산점도
axes[1,0].scatter(df['ICT학교사용'], df['수학점수'], alpha=0.4, color='steelblue')
axes[1,0].set_title('ICT 사용 vs 수학 점수')
axes[1,0].set_xlabel('ICT 학교 사용')
axes[1,0].set_ylabel('수학 점수')

# (4) 유형별 박스플롯
유형들 = ['수동적', '혼합', '능동적']
데이터 = [df[df['ICT활용유형'] == t]['수학점수'] for t in 유형들]
axes[1,1].boxplot(데이터, labels=유형들, patch_artist=True,
                  boxprops=dict(facecolor='lightyellow'))
axes[1,1].set_title('활용 유형별 분포')

plt.suptitle('📊 데이터 탐색 대시보드', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 📝 2편 정리 — 언제 어떤 그래프를 쓰는가?

| 질문 | 그래프 | 코드 |
|------|--------|------|
| 값이 어떻게 퍼져있나? | 히스토그램 | `plt.hist()` |
| 그룹별 평균 차이는? | 막대그래프 | `.plot(kind='bar')` |
| 두 변수가 관련 있나? | 산점도 | `plt.scatter()` |
| 그룹별 분포 비교는? | 박스플롯 | `plt.boxplot()` |
| 비율을 보고 싶다면? | 파이 차트 | `plt.pie()` |
